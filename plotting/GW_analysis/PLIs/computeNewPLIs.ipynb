```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from PLIs import *
from GW_detectors import GW_detectors
import numpy as np
from scipy.interpolate import interp1d
```

```python
N=1000
lisa_pli = PLI_curves(3e-5,0.5,-5,3,-8,8,'lisa',4,10,N).maximise_p()
decigo_pli = PLI_curves(1e-3, 100, -3, 2, -2,8, 'decigo', 4, 10,N).maximise_p()
bbo = PLI_curves(1e-3, 100, -3, 2, -3, 10, 'bbo', 4, 10,N).maximise_p()
bdecigo_pli = PLI_curves(1e-2, 100, -2, 2, -2, 8, 'bdecigo', 4, 8,N).maximise_p()
et_pli = PLI_curves(1, 1e4, 0, 4, -6,6, 'et', 4, 5,N).maximise_p()
```

```python
fig, ax = plt.subplots(figsize = (10,6))
ax.plot(lisa_pli[0],lisa_pli[1])
ax.plot(decigo_pli[0],decigo_pli[1])
ax.plot(bbo[0],bbo[1])
ax.plot(bdecigo_pli[0],bdecigo_pli[1])
ax.plot(et_pli[0],et_pli[1])


ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlim(1e-9, 1e5)

ax.set_ylim(1e-17, 1e-7)
```
